### Scrape Movie Names

```python
url = 'https://timesofindia.indiatimes.com/entertainment/tamil/movie-reviews' 
response = requests.get(url) 
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
movies = soup.find_all('div', class_='pmr_box_Items')

df = pd.DataFrame(columns=['MovieName'])

for movie in movies:
    title = movie.find('div', class_='pmr_box_title').find('h3').text
    df = pd.concat([df, pd.DataFrame({'MovieName': [title]})], ignore_index=True)
```

```python
df.drop_duplicates(inplace=True)
df
```
